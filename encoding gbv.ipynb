```python
# Create ward encoding system for df_encoded
def encode_location_ward(df_encoded):
    """
    Creates numerical encoding for location_ward column
    Maps each ward name to a unique number based on frequency
    """
    
    print("Creating ward encoding system...")
    
    # Get frequency counts of wards
    ward_counts = df_encoded['location_ward'].value_counts()
    
    # Create mapping dictionaries
    ward_to_number = {}
    number_to_ward = {}
    
    # Reserve 0 for Unknown/Missing values
    ward_to_number['Unknown'] = 0
    number_to_ward[0] = 'Unknown'
    
    # Assign numbers 1, 2, 3... to other wards based on frequency (most frequent gets lower numbers)
    current_number = 1
    for ward_name in ward_counts.index:
        if ward_name != 'Unknown' and not pd.isna(ward_name):
            ward_to_number[ward_name] = current_number
            number_to_ward[current_number] = ward_name
            current_number += 1
    
    # Apply encoding to the dataframe
    df_encoded['location_ward_encoded'] = df_encoded['location_ward'].map(ward_to_number).fillna(0)
    
    # Convert to integer type
    df_encoded['location_ward_encoded'] = df_encoded['location_ward_encoded'].astype(int)
    
    print(f" Ward encoding completed!")
    print(f" Total unique wards: {len(ward_to_number)}")
    print(f" Encoded values range: 0 to {max(number_to_ward.keys())}")
    
    # Show sample mappings
    print(f"\n Sample Ward Mappings:")
    print("Code | Ward Name")
    print("-" * 50)
    for i, (ward, freq) in enumerate(ward_counts.head(15).items()):
        code = ward_to_number[ward]
        print(f"{code:4d} | {ward}")
    
    return ward_to_number, number_to_ward

# Apply encoding to your dataframe
ward_to_number, number_to_ward = encode_location_ward(df_encoded)

# Verify the encoding
print(f"\n Verification:")
print(f"Original unique values: {df_encoded['location_ward'].nunique()}")
print(f"Encoded unique values: {df_encoded['location_ward_encoded'].nunique()}")
print(f"Data type: {df_encoded['location_ward_encoded'].dtype}")

# Show sample of encoded data
print(f"\n Sample of encoded data:")
sample = df_encoded[['location_ward', 'location_ward_encoded']].head(10)
print(sample.to_string(index=False))

# Show value counts of encoded column
print(f"\n Top 10 most frequent encoded values:")
encoded_counts = df_encoded['location_ward_encoded'].value_counts().head(10)
for code, count in encoded_counts.items():
    ward_name = number_to_ward.get(code, 'Unknown')
    print(f"Code {code}: {ward_name} ({count} occurrences)")

#print(f"\n Ready for Machine Learning!")
#print(f"Use column: 'location_ward_encoded' in your ML model")
```
