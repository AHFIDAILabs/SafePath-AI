```python
def create_pre_incident_vulnerability_target(df1):
    """
    Create GBV Vulnerability Target Using ONLY Pre-Incident Characteristics
    
    Uses demographic, socioeconomic, and social factors that exist BEFORE GBV occurs
    to predict vulnerability for early warning and prevention.
    
    Based on Rohingya refugee research showing key pre-incident vulnerability factors:
    - Demographics (gender, age, marital status)  
    - Economic status (employment, education)
    - Social vulnerability (living situation, special populations)
    - Support systems availability
    
    Target: 1 = Vulnerable to GBV, 0 = Non-Vulnerable to GBV
    """
    
    # Initialize vulnerability risk assessment
    vulnerability_score = np.zeros(len(df1))
    
    # ================================================================
    # 1. DEMOGRAPHIC PRE-INCIDENT RISK FACTORS (Weight: 40%)
    # Journal: "Women & girls = 52% of vulnerable population"
    # ================================================================
    
    demographic_risk = np.zeros(len(df1))
    
    # Gender Risk - Primary vulnerability factor
    # Journal shows females significantly more vulnerable
    gender_risk = (df1['survivor_sex'] == 'Female').astype(float)
    
    # Age-Based Vulnerability Categories
    age_risk = np.zeros(len(df1))
    
    # Children and adolescents (0-17) - High vulnerability
    # Journal: "Nearly 100% of women/girls experienced or witnessed GBV"
    children_risk = ((df1['survivor_age'] >= 0) & (df1['survivor_age'] <= 17)).astype(float) * 0.8
    
    # Young women (18-30) - Peak vulnerability period
    young_women_risk = ((df1['survivor_age'] >= 18) & (df1['survivor_age'] <= 30) & 
                       (df1['survivor_sex'] == 'Female')).astype(float) * 0.7
    
    # Elderly (60+) - Physical vulnerability, isolation
    elderly_risk = (df1['survivor_age'] >= 60).astype(float) * 0.6
    
    age_risk = np.maximum.reduce([children_risk, young_women_risk, elderly_risk])
    
    # Marital Status Risk
    # Journal: "1/6 families headed by single mothers"
    marital_risk = np.zeros(len(df1))
    marital_risk += (df1['marital_status'] == 'Single').astype(float) * 0.6
    marital_risk += (df1['marital_status'] == 'Divorced/separated').astype(float) * 0.8
    marital_risk += (df1['marital_status'] == 'Widowed').astype(float) * 0.9
    marital_risk += (df1['marital_status'] == 'Unknown').astype(float) * 0.4
    
    # Combine demographic factors
    demographic_risk = (gender_risk * 0.4 + age_risk * 0.4 + marital_risk * 0.2)
    vulnerability_score += demographic_risk * 0.40
    
    # ================================================================
    # 2. ECONOMIC VULNERABILITY (Weight: 25%)
    # Journal: "45% of female-headed households vulnerable to food insecurity"
    # ================================================================
    
    economic_risk = np.zeros(len(df1))
    
    # Employment Status - Key economic vulnerability indicator
    employment_risk = np.zeros(len(df1))
    employment_risk += (df1['employment_status_victim_main'] == 'Unemployed').astype(float) * 1.0
    employment_risk += (df1['employment_status_victim_main'] == 'Unknown').astype(float) * 0.7
    employment_risk += (df1['employment_status_victim_main'] == 'Self employed').astype(float) * 0.4  # Economic instability
    employment_risk += (df1['employment_status_victim_main'] == 'Not reported').astype(float) * 0.6
    
    # Educational Vulnerability
    # Lower education increases economic dependency and reduces options
    education_risk = np.zeros(len(df1))
    education_risk += (df1['educational_status'] == 'No formal').astype(float) * 1.0
    education_risk += (df1['educational_status'] == 'Unknown').astype(float) * 0.7
    education_risk += (df1['educational_status'] == 'Some primary').astype(float) * 0.8
    education_risk += (df1['educational_status'] == 'Completed primary').astype(float) * 0.6
    education_risk += (df1['educational_status'] == 'Some secondary').astype(float) * 0.4
    
    # Combine economic factors
    economic_risk = (employment_risk * 0.6 + education_risk * 0.4)
    vulnerability_score += economic_risk * 0.25
    
    # ================================================================
    # 3. SOCIAL VULNERABILITY & ISOLATION (Weight: 20%)
    # Journal: "Absence of women-friendly spaces - most significant driver"
    # ================================================================
    
    social_risk = np.zeros(len(df1))
    
    # Living Situation Risk - Critical protection factor
    living_risk = np.zeros(len(df1))
    living_risk += (df1['who_survivor/victim_stay_with'] == 'Alone').astype(float) * 1.0  # Highest risk
    living_risk += (df1['who_survivor/victim_stay_with'] == 'Unknown').astype(float) * 0.8
    living_risk += (df1['who_survivor/victim_stay_with'] == 'Other').astype(float) * 0.6
    
    # Special risk for minors with guardians (potential abuse situations)
    guardian_minor_risk = ((df1['who_survivor/victim_stay_with'] == 'Guardian') & 
                          (df1['survivor_age'] < 18)).astype(float) * 0.7
    
    social_risk = np.maximum(living_risk, guardian_minor_risk)
    vulnerability_score += social_risk * 0.20
    
    # ================================================================
    # 4. SPECIAL POPULATION VULNERABILITIES (Weight: 15%)
    # Journal: "Overlapping vulnerabilities compound risk"
    # ================================================================
    
    special_population_risk = np.zeros(len(df1))
    
    # High-risk populations based on journal research
    special_population_risk += (df1['PLHIV'] == True).astype(float) * 0.3  # Health vulnerability
    special_population_risk += (df1['IDP'] == True).astype(float) * 0.4   # Displacement = higher risk
    special_population_risk += (df1['widow'] == True).astype(float) * 0.4  # Social/economic vulnerability
    special_population_risk += (df1['minor'] == True).astype(float) * 0.3  # Age vulnerability
    special_population_risk += (df1['out_of_school_child'] == True).astype(float) * 0.4  # Lack of protection
    special_population_risk += (df1['orphans'] == True).astype(float) * 0.5  # No family protection
    special_population_risk += (df1['drug_user'] == True).astype(float) * 0.3  # Additional vulnerability
    special_population_risk += (df1['household_help'] == True).astype(float) * 0.4  # Power imbalance
    special_population_risk += (df1['female_sex_worker'] == True).astype(float) * 0.5  # High-risk occupation
    
    # Prevent over-scoring for multiple vulnerabilities
    special_population_risk = np.minimum(special_population_risk, 1.0)
    vulnerability_score += special_population_risk * 0.15
    
    # ================================================================
    # 5. CREATE BINARY VULNERABILITY TARGET
    # ================================================================
    
    # Normalize vulnerability score to 0-1 range
    vulnerability_score = np.clip(vulnerability_score, 0, 1)
    
    # Set vulnerability threshold based on journal research
    # Journal shows approximately 30% of population at high vulnerability
    # Using 70th percentile to identify those at highest risk
    vulnerability_threshold = np.percentile(vulnerability_score, 70)
    
    # Create binary target variable
    # 1 = Vulnerable to GBV, 0 = Non-Vulnerable to GBV
    vulnerability_target = (vulnerability_score >= vulnerability_threshold).astype(int)
    
    return vulnerability_target
```

```python
df1['vulnerability_target'] = create_pre_incident_vulnerability_target(df1)
```

```python
import os
import seaborn as sns
```

```python
# Create the save folder if not exists
save_path = r"C:\Users\hp\Desktop\gbv_charts"

os.makedirs(save_path, exist_ok=True)

# Helper function to save + show plots
def save_and_show(fig_name):
    plt.savefig(os.path.join(save_path, fig_name), bbox_inches="tight")
    plt.show()
```

```python

# ----------------------------------------------------
# 1. Demographic Profile of Survivors
# ----------------------------------------------------
sns.countplot(x='survivor_sex', data=df1)
plt.title("Gender Distribution of Survivors")
save_and_show("1_gender_distribution.png")

sns.histplot(df1['survivor_age'], bins=30, kde=True)
plt.title("Age Distribution of Survivors")
plt.xlabel("Age")
save_and_show("1_age_distribution.png")

sns.countplot(y='marital_status', data=df1, order=df1['marital_status'].value_counts().index)
plt.title("Marital Status of Survivors")
save_and_show("1_marital_status.png")

sns.boxplot(x='survivor_sex', y='survivor_age', data=df1)
plt.title("Age Distribution by Gender")
save_and_show("1_age_vs_gender.png")

print("\n Marital Status vs vulnerability_target:")
print(pd.crosstab(df1['marital_status'], df1['vulnerability_target'], normalize="index")*100)
```
