```python
def determine_vulnerability_safe(row):
    """
    Determines vulnerability label from ONLY pre-incident variables.
    Avoids leakage by excluding anything that can only be known after the incident.
    """

    score = 0
    reasons = []

    # 1️⃣ Age factor
    age = row.get('age_of_survivor', None)
    if pd.notna(age):
        if age < 18:
            score += 3
            reasons.append('Minor (<18)')
        elif age >= 65:
            score += 2
            reasons.append('Elderly (65+)')

    # 2️⃣ Marital status factor
    marital = row.get('marital_status', '').strip().lower()
    if marital in ['single', 'divorced/separated', 'widowed']:
        score += 1
        reasons.append('Marital status risk')

    # 3️⃣ Living arrangement factor
    living = row.get('who_survivorvictim_live_with', '').strip().lower()
    if living in ['alone', 'perpetrator/trafficker', 'idp camp/shelter']:
        score += 2
        reasons.append('Living arrangement risk')

    # 4️⃣ Employment factor
    emp_status = row.get('survivorvictim_employment_status', '').strip().lower()
    if emp_status in ['unemployed', 'housewife', 'retired']:
        score += 1
        reasons.append('Employment status risk')

    # 5️⃣ Vulnerable population flag
    vuln_pop = row.get('vulnerable_population', '').strip().lower()
    if vuln_pop not in ['', 'no vulnerability', 'unknown']:
        score += 3
        reasons.append('Listed vulnerable population')

    # Final classification
    if score >= 5:
        label = 'Highly Vulnerable'
    elif score >= 3:
        label = 'Vulnerable'
    else:
        label = 'Less Vulnerable'

    return {'label': label, 'score': score, 'reasons': reasons}
```

```python
# Apply the function to each row
vulnerability_results = df1.apply(determine_vulnerability_safe, axis=1)

# Convert the list of dicts to separate columns
vulnerability_df = pd.DataFrame(vulnerability_results.tolist())

# Merge with the original dataframe
df1 = pd.concat([df1, vulnerability_df], axis=1)
```
